## Encode Categorical Variables

```python
# Create a function 

def create_dummy_df(df, cat_cols, dummy_na):

'''

Description: This function encodes dummy variables from specified columns in a given dataset.

Arguments: 
    df: the dataframe
    cat_cols: the list of columns to encode into dummy variables
    dummy_na: boolean indicates whether or not to create a dummy variable to indicate NA values in a given column

Returns: the same dataframe with original variables in cat_cols dropped and replaced with dummy variable columns

'''
    
    
    for col in cat_cols:
        
        try:
            df = pd.concat([df.drop(col, axis=1), 
                            pd.get_dummies(df[col], 
                                           prefix=col, 
                                           prefix_sep='_', 
                                           drop_first=True, dummy_na=dummy_na)], axis=1)
        except:
            continue
    
    return df
```

```python
# Create a new dataframe with dummy variables
cat_copy = create_dummy_df(cat_vars, cat_vars.columns, dummy_na=False)
```

# Data prep for Question 2: Ratings Prediction

Next we'll look at the question of whether we can generate a model that predicts ratings with a reasonable degree of accuracy. First we need to do some data cleaning and explore which variables to use for this model.

```python
#select just the columns we plan to use for this model into a new df

ratings_model_df = df_lst[['id', 'review_scores_rating', 'neighbourhood_group_cleansed', 'reviews_per_month', 
                            'price', 'property_type', 'host_response_time', 'host_is_superhost', 'room_type', 
                            'host_response_rate', 'cleaning_fee', 'host_listings_count', 'security_deposit']].copy()
```

```python
#check for nas

ratings_model_df.isna().apply(sum) / len(ratings_model_df) # % of nas in each column
```

```python
#convert data types
ratings_model_df['security_deposit'] = ratings_model_df['security_deposit'].str.replace('$', '').str.replace(',','').astype(float)
ratings_model_df['host_response_rate'] = ratings_model_df['host_response_rate'].str.replace('%', '').astype(float)
ratings_model_df['cleaning_fee'] = ratings_model_df['cleaning_fee'].str.replace('$', '').str.replace(',','').astype(float)
ratings_model_df['price'] = ratings_model_df['price'].str.replace('$', '').str.replace(',','').astype(float)
```

```python
ratings_model_df['host_is_superhost'] = ratings_model_df['host_is_superhost'].replace({'t': 1, 'f': 0})
```

```python
#fill nas
ratings_model_df['security_deposit'] = ratings_model_df['security_deposit'].fillna(0)
ratings_model_df['cleaning_fee'] = ratings_model_df['cleaning_fee'].fillna(0)
ratings_model_df['host_response_rate'] = ratings_model_df['host_response_rate'].fillna(0)
```

```python
#check output
ratings_model_df[['cleaning_fee', 'security_deposit', 'host_response_rate', 'host_is_superhost']]
```

```python
#creating a dictionary of amenities with listing ids 
amen_dict = {}

for i in range(len(df_lst['amenities'])):
    items = df_lst['amenities'][i].split(",")
    
    for item in items:
        
        item = item.strip("{").strip("}").strip("\"").strip(" ")
        
        if item in amen_dict:
            amen_dict[item].append(df_lst['id'][i])
        else:
            amen_dict[item] = [df_lst['id'][i]]
```

```python
#check dictionary keys
amen_dict.keys()
```

```python
def dummy_var_amenity(df, dict_name, df_new_col, dict_key):
    
'''
Description: Creates a new dummy variable in a dataset by searching a dictionary for each row's unique id and
returning 1 if the id is found in the dictionary's values

Arguments:
    df: the dataframe
    dict_name: the dictionary containing lists of unique ids from the dataframe in the values
    df_new_col: the name of the new dummy variable to be added to the dataframe
    dict_key: the dictionary key for the values that should be searched

Returns:
    None. Modifies the dataframe in place

'''
    
    #create a new dummy var column for the amenity, with 0 as default
    df.insert(df.shape[1], df_new_col, np.zeros(len(df), dtype=int), True)
    
    #put a 1 in the column if the listing id is associated with the amenity
    for i in df.index:
        if df['id'][i] in dict_name[dict_key]:
            df[df_new_col][i] = 1
```

```python
#create dummy variables from amenities dictionary
dummy_var_amenity(ratings_model_df, amen_dict, 'free_parking', 'Free Parking on Premises')
dummy_var_amenity(ratings_model_df, amen_dict, 'washer', 'Washer')
dummy_var_amenity(ratings_model_df, amen_dict, 'dryer', 'Dryer')
dummy_var_amenity(ratings_model_df, amen_dict, 'kitchen', 'Kitchen')
dummy_var_amenity(ratings_model_df, amen_dict, 'pets', 'Pets Allowed')
```

```python
#check for any remaining nas
ratings_model_df.dropna(subset=['review_scores_rating']).isna().apply(sum) / len(ratings_model_df.dropna(subset=['review_scores_rating']))
```

Dropping NAs in the `reviews_scores_rating` helps solve some missing data problems in other variables. We'll drop the missing data in the `review_scores_rating` since we can't predict the data for missing values.

```python
#drop nas in ratings and predictor variables
ratings_model_df = ratings_model_df.dropna(subset=['review_scores_rating', 'host_response_time'])
```

```python
#turn categorical variables into dummy variables

ratings_model_df = pd.get_dummies(ratings_model_df, dummy_na = True)
```

```python
#run a correlation matrix to find relationships
corr = ratings_model_df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(20, 18))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Looks like there aren't that many that are closely related.

Looks like our most highly related variable is `host_is_superhost`. Variables related to response time and response rate are also fairly related, as well as number of reviews the listing has per month. The `host_listings_count` and being in the University District neighborhood seem to be negatively related. Surprisinly, `security_deposit`, `cleaning_fee`, and even `price` don't appear to be giving us a lot of information about ratings.
